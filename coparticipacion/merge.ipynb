```python
import os
import pandas as pd
years = [2015,2016,2017,2018, 2019, 2020, 2021, 2022, 2023, 2024]

cwd = os.getcwd()
directorios = []
data = pd.DataFrame()
for year in years:
    folder = os.path.join(cwd, str(year))
    for carpeta_raiz, subcarpetas, archivos in os.walk(folder):
        for archivo in archivos:
            if archivo.startswith(("t","internet_diario")):
                directorios.append([os.path.join(carpeta_raiz,archivo), os.path.basename(carpeta_raiz), archivo])
                try:
                    dt = pd.read_excel(os.path.join(carpeta_raiz,archivo), sheet_name="t")
                except:
                    dt = pd.read_excel(os.path.join(carpeta_raiz,archivo), sheet_name="Distribución en Pesos")
                
                columna_dias = dt.columns[0]
                dt = dt[dt[columna_dias].isin(range(1,32))]
                dt["periodo"] = os.path.basename(carpeta_raiz)
                
                data = pd.concat([data, dt],axis=0)
                
df = pd.DataFrame(directorios, columns=['Carpeta Raíz', 'Subcarpeta', 'Archivo'])
    
```

```python

data.reset_index(drop=True,inplace=True)
drop = data[data["Unnamed: 28"]==0].index

data.drop(drop,axis=0,inplace=True)
data["year"] = data["periodo"].str[-4:]
data["month"] = data["periodo"].str[:-4]
meses = {'enero':1, 'febrero':2, 'marzo':3, 'abril':4,
         'mayo':5, 'junio':6, 'julio':7, 'agosto':8,
         'septiembre':9, 'octubre':10, 'noviembre':11, 'diciembre':12}
data["date"] = data["Unnamed: 0"].astype(str) + "/" + data["month"].map(meses).astype(str) + "/" + data["year"].astype(str)
data["date"] = pd.to_datetime(data["date"], format="%d/%m/%Y")
data["date"] = data["date"].dt.strftime("%d/%m/%Y")
data.head()
```
