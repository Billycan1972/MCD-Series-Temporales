<h3>Pregunta 1= Describa las caracteristicas principales de un modelo autoregresivo</h3>

1. Dependencia de valores pasados: Un modelo AR utiliza los valores anteriores de la serie temporal para predecir los futuros. Esto significa que los valores futuros son una función lineal de los valores anteriores y un término de error aleatorio.

2. Orden del modelo: El orden de un modelo AR, denotado como AR(p), indica el número de valores pasados utilizados en la predicción. Por ejemplo, un modelo AR(1) usa el valor inmediatamente anterior, mientras que un AR(2) usa los dos valores anteriores.
3. Estacionariedad: Los modelos AR son más efectivos con series temporales estacionarias, lo que significa que las propiedades estadísticas (como la media y la varianza) son constantes en el tiempo.

<h3>Pregunta 2= Describa las caracteristicas principales de un modelo de medias moviles</h3>

1. Dependencia de errores pasados: A diferencia del modelo AR, un modelo MA utiliza los errores (residuos) pasados para predecir el valor futuro. Estos errores representan la parte impredecible de la serie.

2. Orden del modelo: El modelo MA(q) tiene un orden q que indica el número de errores pasados utilizados en la predicción. Por ejemplo, un MA(1) usa el error del periodo anterior, mientras que un MA(2) usa los errores de los dos periodos anteriores.
3. Estacionariedad: Los modelos MA son siempre estacionarios, ya que no dependen de los valores pasados de la serie, sino de los errores, que se asumen como ruido blanco.

<h3>Pregunta 3= Describa las caracteristicas principales de un modelo de autoregresivo de medias moviles</h3>

1. Combinación de AR y MA: Un modelo ARMA combina las características de los modelos autoregresivos (AR) y de medias móviles (MA), es decir, usa tanto los valores pasados de la serie como los errores pasados para hacer predicciones.

2. Orden del modelo: Un modelo ARMA tiene dos órdenes, ARMA(p, q), donde
𝑝
p es el número de términos autoregresivos y
𝑞
q es el número de términos de medias móviles.
3. Estacionariedad: Al igual que el modelo AR, un ARMA necesita que la serie sea estacionaria. Si la serie no es estacionaria, puede ser necesario aplicar transformaciones como la diferenciación.

<h3>Pregunta 4= ¿ cual es el criterio utilizado mas destacado para determinar el orden optimo de un modelo AR, MA o ARMA.?</h3>

El criterio más destacado y ampliamente utilizado para determinar el orden óptimo de un modelo AR, MA, o ARMA es el Criterio de Información de Akaike (AIC). Sin embargo, también se utilizan otros criterios como el Criterio de Información Bayesiano (BIC) y el Criterio de Hannan-Quinn (HQIC). A continuación, te explico los más importantes:


<b>Criterio de Información de Akaike (AIC):</b> <br>
Descripción: El AIC es una medida que evalúa la calidad de un modelo en función de su capacidad de ajuste (verosimilitud) y su simplicidad (penaliza el número de parámetros). El objetivo es encontrar el modelo con el menor valor de AIC.

<b>Criterio de Información Bayesiano (BIC):</b> <br>
Descripción: El BIC es similar al AIC, pero penaliza de manera más fuerte el número de parámetros. Esto significa que BIC tiende a seleccionar modelos más simples en comparación con AIC.

<b>Criterio de Hannan-Quinn (HQIC) </b><br>
Descripción: Es otro criterio similar a AIC y BIC, pero tiene una penalización intermedia, menos fuerte que BIC y más fuerte que AIC.

<b>Comparación y Elección:</b><br>
* AIC tiende a seleccionar modelos más complejos (mayor número de parámetros), ya que penaliza de manera más suave.
* BIC es más conservador, prefiriendo modelos más simples, por lo que es más adecuado cuando se tiene un conjunto de datos grandes y se quiere evitar el sobreajuste.
* HQIC es una alternativa que penaliza de manera más fuerte que AIC, pero menos que BIC, y a veces puede ser útil en casos donde ni AIC ni BIC funcionan bien.

En resumen, el AIC es el criterio más destacado y comúnmente utilizado, pero en contextos donde se desea un mayor control sobre la complejidad del modelo, el BIC también es una excelente opción.

<h3>¿Como determinar el orden p, d, q de un modelo AR, MA y/o ARMA?</h3>

Para determinar estos órdenes, se siguen generalmente estos pasos:

1. Análisis gráfico:
  * Gráfico de la serie temporal
  * Función de Autocorrelación (ACF)
  * Función de Autocorrelación Parcial (PACF)


2. Criterios de información:
  * AIC (Criterio de Información de Akaike)
  * BIC (Criterio de Información Bayesiano)


3. Pruebas de diagnóstico:
  * Análisis de residuos
  * Pruebas de normalidad
  * Pruebas de autocorrelación de residuos



El proceso típico es:

1. Determinar d: Se analiza la estacionariedad de la serie. Si no es estacionaria, se diferencia hasta lograr la estacionariedad.

2. Determinar p y q:
  * Para AR: Se observa la PACF. El número de rezagos significativos sugiere el orden p.
  * Para MA: Se observa la ACF. El número de rezagos significativos sugiere el orden q.
  * Para ARMA: Se comparan varios modelos con diferentes combinaciones de p y q.


3. Selección del modelo:

  * Se ajustan varios modelos con diferentes órdenes.
  * Se comparan utilizando AIC o BIC. El modelo con el valor más bajo de estos criterios suele considerarse el mejor.


3. Validación del modelo:

  * Se realizan pruebas de diagnóstico para asegurar que el modelo es adecuado.

<h3>Pregunta 5= ¿Cuál es la razón para optar por el principio de la parsimonia en la toma de decisiones para la seleccion de modelos ? </h3>

El principio de parsimonia, también conocido como la navaja de Occam, es fundamental en la selección de modelos estadísticos, incluidos los modelos AR, MA y ARMA. La razón principal para optar por este principio en el contexto de estos modelos se basa en varios aspectos importantes:

1. Evitar el sobreajuste:
Un modelo más complejo (con más parámetros) tiende a ajustarse mejor a los datos de entrenamiento, pero puede "aprender" el ruido en los datos en lugar de la verdadera estructura subyacente. Esto puede llevar a un rendimiento deficiente en nuevos datos.

2. Mejor generalización:
Los modelos más simples suelen generalizar mejor a nuevos datos, lo que es crucial para hacer predicciones precisas.
3. Interpretabilidad:
Los modelos más simples son generalmente más fáciles de interpretar y explicar, lo que es importante en muchos contextos de toma de decisiones.
4. Estabilidad:
Los modelos más simples tienden a ser más estables y menos sensibles a pequeños cambios en los datos de entrada.
5. Eficiencia computacional:
Modelos más simples requieren menos recursos computacionales para ser estimados y utilizados.
6. Reducción del error de estimación:
Cada parámetro adicional introduce un error de estimación. Al minimizar el número de parámetros, se reduce este error acumulativo.
7. Balance entre sesgo y varianza:
Los modelos más simples pueden tener un mayor sesgo pero menor varianza, lo que a menudo resulta en un mejor rendimiento general.
8. Criterios de información:
Los criterios como AIC y BIC incluyen penalizaciones por la complejidad del modelo, favoreciendo así modelos más simples.

En el contexto específico de los modelos AR, MA y ARMA:

* Para un modelo AR, preferir un orden p más bajo significa asumir que los valores pasados más recientes son los más influyentes.
* Para un modelo MA, un orden q más bajo implica que solo los errores más recientes son significativos para predecir el valor actual.
* En modelos ARMA, buscar la combinación más simple de p y q que capture adecuadamente la dinámica de la serie temporal.

Es importante notar que la parsimonia no significa siempre elegir el modelo más simple, sino encontrar un equilibrio entre la simplicidad y la capacidad explicativa. El objetivo es capturar la estructura esencial de los datos sin incluir complejidad innecesaria.

<h3>Pregunta 6= ¿Cuáles son las dos propiedades clave que deben tener los residuos para que un método de pronóstico se considere bueno según el texto? </h3> - Puede tomar como referencia el capitulo 3 del libro recomendado de Hyndman.

Basándome en el capítulo 3 del libro de Hyndman sobre pronósticos, las dos propiedades clave que deben tener los residuos para que un método de pronóstico se considere bueno son:

1. Los residuos deben ser no correlacionados.
2. Los residuos deben tener media cero.

Estas propiedades son fundamentales por las siguientes razones:

1. Residuos no correlacionados:

  * Significa que los residuos no deben mostrar ningún patrón.
  * Si hay correlación en los residuos, indica que hay información en los errores que debería ser utilizada en el pronóstico.
  * La ausencia de correlación sugiere que toda la información útil ha sido extraída de los datos.


2. Residuos con media cero:

  * Implica que los pronósticos no están sesgados.
  * Si los residuos tienen una media diferente de cero, los pronósticos están sistemáticamente muy altos o muy bajos.



Estas propiedades se pueden verificar mediante:

* Gráficos de los residuos a lo largo del tiempo para detectar patrones.
* Gráficos de autocorrelación (ACF) de los residuos para comprobar la correlación.
* Pruebas estadísticas como el test de Ljung-Box para la autocorrelación.
* Cálculo de la media de los residuos y pruebas t para verificar si es significativamente diferente de cero.

Es importante notar que, aunque Hyndman destaca estas dos propiedades como las más importantes, en la práctica también se suelen considerar otras características de los residuos, como la normalidad y la homocedasticidad (varianza constante), especialmente para ciertos tipos de modelos y pruebas estadísticas.

```python
# Aqui hacemos el loop probando diferentes "p" y "q", para saber cual es el criterio de informacion mas bajo.

import itertools
import statsmodels.api as sm

p_values = range(0, 3)
q_values = range(0, 3)


resultados = []
mejor_aic = float("inf")
mejor_order = None
mejor_modelo = None


for p, q in itertools.product(p_values, q_values):
    try:

        modelo = sm.tsa.arima.ARIMA(df, order=(p, 0, q))
        resultado = modelo.fit()

        aic = resultado.aic
        resultados.append((p, q, aic))

        if aic < mejor_aic:
            mejor_aic = aic
            mejor_order = (p, q)
            mejor_modelo = resultado

    except Exception as e:
        print(f"Error en el modelo ARMA({p},{q}): {e}")

print(f"Mejor modelo ARMA: order={mejor_order} con AIC={mejor_aic}")

for p, q, aic in resultados:
    print(f"ARMA({p},{q}) AIC: {aic}")
```
