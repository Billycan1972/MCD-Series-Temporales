```python
import pytimetk as tk
import pandas as pd

df = tk.load_dataset('bike_sales_sample')
df['order_date'] = pd.to_datetime(df['order_date'])

df
```

La función summarize_by_time() se utiliza para resumir datos de series temporales. En el contexto de un análisis de ventas, se puede usar para calcular el total de ingresos de bicicletas de montaña y de carretera por mes

```python
summary_category_1_df = df \
    .groupby("category_1") \
    .summarize_by_time(
        date_column  = 'order_date',
        value_column = 'total_price',
        freq         = "MS",
        agg_func     = 'sum',
        wide_format  = False
    )

summary_category_1_df.head()
```

Visualizando Patrones de Ventas


```python
summary_category_1_df \
    .groupby('category_1') \
    .plot_timeseries(
        date_column  = 'order_date',
        value_column = 'total_price',
        smooth_frac  = 0.8
    )
```

Detección de anomalías

**m4_monthly**: muestra de 4 series de tiempo de frecuencia mensual tomadas de la M4 competition.

A continuación cargamos los datos, seleccionamos in ID de interés y visualizamos la serie:

```python
# Librerias
import pytimetk as tk
import pandas as pd
import numpy as np

# Importamos datos
m4_daily_df = tk.load_dataset('m4_daily', parse_dates = ['date'])
```

```python
# Filtramos datos
df = (
    m4_daily_df
        .query("id == 'D10'")
        .query("date.dt.year == 2015")
)
```

```python
# Grafico de datos
tk.plot_timeseries(
    data         = df,
    date_column  = 'date',
    value_column = 'value'
)
```

**iqr_alpha** controla el umbral para detectar valores atípicos. Es el nivel de significancia utilizado en el método del rango intercuartil (IQR) para la detección de valores atípicos. El valor predeterminado es 0,05, que corresponde a un nivel de significancia del 5%. Un nivel de significancia más bajo dará como resultado un umbral más alto, lo que significa que se detectarán menos valores atípicos. Un nivel de significancia más alto dará como resultado un umbral más bajo, lo que significa que se detectarán más valores atípicos.


```python
# Anomalize
anomalize_df = tk.anomalize(
    data          = df,
    date_column   = 'date',
    value_column  = 'value',
    period        = 7, #  con período de estacionalidad en días
    iqr_alpha     = 0.05, # valor por defecto
    clean_alpha   = 0.75, # valor por defecto
    clean         = "min_max"
)

anomalize_df.glimpse()
```

Visualizamos la seaonal decomposition para ver su representación:

```python
# grafico de descoposicion estacional
tk.plot_anomalies_decomp(
    data        = anomalize_df,
    date_column = 'date',
    engine      = 'plotly',
    title       = 'Seasonal Decomposition'
)
```

Visualizamos las anomalías ustilizando tk.plot_anomalies();

```python
# grafico de anomalias
tk.plot_anomalies(
    data        = anomalize_df,
    date_column = 'date',
    engine      = 'plotly',
    title       = 'Plot Anomaly Bands'
)
```

Finalmente observamos la grafic con los datos limpios utilizando plot_anomalies_cleaned()

```python
# Trazar anomalías limpiadas
tk.plot_anomalies_cleaned(
    data        = anomalize_df,
    date_column = 'date'
)
```

```python
# Datos anomalizados con múltiples valores iqr_alpha

# valores de Alpha
iqr_alpha_values = [0.05, 0.10, 0.15, 0.20]


dfs = []

for alpha in iqr_alpha_values:


    anomalize_df = tk.anomalize(
        data         = df,
        date_column  = 'date',
        value_column = 'value',
        period       = 7,
        iqr_alpha    = alpha
    )

    # - Agrega la columna iqr_alpha
    anomalize_df['iqr_alpha'] = f'iqr_alpha value of {alpha}'


    dfs.append(anomalize_df)

final_df = pd.concat(dfs)
```

```python

(
    final_df
        .groupby('iqr_alpha')
        .plot_anomalies(
            date_column = 'date',
            engine      = 'plotly',
            facet_ncol  = 2
        )
)
```
